```python
from a1_part2_m2 import (
    CustomCNN, train_model, test_model,
    AircraftDataset, transform, train_df, test_df, device
)
from torch.utils.data import DataLoader

# Setup
train_dataset = AircraftDataset(train_df, transform=transform)
test_dataset = AircraftDataset(test_df, transform=transform)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
num_classes = max(train_dataset.df['label']) + 1

# Train and evaluate
model = CustomCNN(num_classes=num_classes).to(device)
train_model(model, train_loader, num_epochs=5)
accuracy = test_model(model, test_loader)

# Display accuracy in notebook
print(f"Final Test Accuracy of Custom CNN: {accuracy:.2f}%")
```

- We tested two methods: a pretrained ResNet-50 model and a custom CNN with residual connections.
- For ResNet-50:
  - **Transfer Learning** (5 epochs, frozen layers): achieved **29.40%** test accuracy.
  - **Fine-Tuning** (5 epochs, all layers trainable): achieved **50.62%** test accuracy.
- Our **Custom CNN** (5 epochs, trained from scratch with residual blocks) achieved **2.10 %** accuracy, showing limited performance on complex fine-grained classification.
- ResNet-50 with fine-tuning outperformed other methods, highlighting the value of pretrained features for small and detailed datasets.

**Augmentation Strategy Implemented:**

- Applied `RandomHorizontalFlip` with 0.5 probability
- Used `ColorJitter` with brightness and contrast adjustments up to 0.3
- Normalized all images with mean and std of 0.5 per channel

**Learning Rate Scheduler Used:**
- Used `StepLR` with `step_size=5` and `gamma=0.1`
- This reduced the learning rate by a factor of 10 after every 5 epochs to help refine the learning phase

**Performance Results:**
- Baseline Custom CNN Accuracy: 2.67%
- With Learning Rate Scheduler: 2.97%
- With Augmentation + LR Scheduler: **3.87%**

**Effectiveness Analysis:**
- The augmentation strategy introduced controlled randomness that helped the model generalize better to new samples.
- Resizing and cropping simulated slight viewpoint variation, while flipping and color jitter mimicked real-world lighting conditions and orientations.
- The learning rate scheduler helped prevent overfitting and enabled the optimizer to fine-tune weights as training progressed.
- Overall, the combination of augmentation and scheduler increased robustness and learning efficiency.
- **Performance improved by +1.20%** over the baseline, indicating the impact of both techniques working together.

### Limitations of the Two Algorithms

#### ResNet-50 (Pretrained Model)
- The pretrained ResNet-50 model performed significantly better than the custom CNN (~37% accuracy), but it comes with trade-offs.
- Being a deep architecture with over 23 million parameters, it is computationally expensive, especially when fine-tuning all layers.
- The pretrained weights were learned from ImageNet, which contains generic object classes. This introduces a domain gap when applied to fine-grained tasks like distinguishing between similar aircraft variants.
- The model tends to overfit quickly if the entire network is fine-tuned without proper regularization or scheduling.
- It requires substantial GPU memory and training time, making it less practical for lightweight or real-time applications.

#### Custom CNN (Self-Designed Model)
- Despite structural improvements like residual blocks and increased depth, the model achieved only ~3.87% accuracy, indicating limited capacity in capturing fine-grained differences across classes.
- Without any pretrained features, the model had to learn from scratch, which is a major disadvantage given the relatively small size and imbalance of the training set.
- The absence of global feature abstraction (like adaptive pooling or multi-scale context) limits its ability to generalize.
- It struggles to converge effectively on complex classes due to insufficient representation learning in deeper layers.

---

### Potential Improvements

#### ResNet-50 (Pretrained Model)
- Freeze early layers during initial training to reduce overfitting and speed up convergence, then selectively fine-tune deeper layers.
- Combine with custom classification heads designed specifically for this task, such as those using attention mechanisms or fine-grained loss functions (e.g., focal loss).
- Utilize domain-adaptive pretraining (e.g., pretraining on aviation-specific datasets) to reduce the domain shift from ImageNet to FGVC-Aircraft.

#### Custom CNN (Self-Designed Model)
- Add Squeeze-and-Excitation (SE) blocks or attention layers to help the model focus on important features.
- Incorporate deeper residual paths and dilated convolutions to improve context awareness.
- Use hybrid approaches—initialize with pretrained encoders from smaller models (like MobileNet or EfficientNet-B0) and fine-tune the final blocks.
- Increase model robustness with dropout regularization and better architectural tuning like bottleneck layers or grouped convolutions.
